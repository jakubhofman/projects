### __FEATURE SELECTION FOR MACHINE LEARNING__


_Attention : in order to re-run notebook you need to put DLBCL and Prostate files together with notebook in the same directory_

Research conducted with following algorithms: 

1.  Scoring fucntion: ANOVA
2.  SVM classifier with recursive feature elimination and crossvalidation 
3.  Decision Tree and RandomForest classifiers 

Parameter modified in the study was number of features. Parameter was included in the algoritms as follows:
1. Anova : returns number of features with the highest numbers of points
2. RFE SVM : number of features means minimum number of features returned by aglorithm ("depth of analysis")
3. Decision trees : maximum number of featuresmaksymalna to be included in tree split

Study conducted for the following parameters:   
number of features = 0,1%,0.5%,1%,5%,10%,15%,20%,30%,50% of all features in the datasets.

Results of the algorithms, which is quality of the selected feature set, was measured with mean square error of classification using:
NBC,5NN,J48. Additionaly computation time was measured.
 



```python
import selekcja_cech as sc
import pandas as pd
```

```python
procent_atr=[0.1,0.5,1,5,10,15,20,30,50]
dlbcl_training = 'dlbcl_training.csv'
dlbcl_test ='dlbcl_testing.csv'

pro_training = 'prostate_training.csv'
pro_test ='prostate_testing.csv'


X_train,y_train,X_test,y_test = sc.load_data(dlbcl_training,dlbcl_test)
dlbcl=sc.select_feature(X_train, y_train, X_test, y_test, procent_atr)

X_train,y_train,X_test,y_test = sc.load_data(pro_training,pro_test)
prostate=sc.select_feature(X_train, y_train, X_test, y_test, procent_atr)
```

### 1. __RESULTS__

#### 1.1 Best selected features set:<br>
    Due to the identical results of selection quality('Error') second criterion is the samllest size of the feature set('N.features')

DLBCL : 

```python
df=dlbcl.sort_values(by=['Błąd','L.cech'])
df.columns=['Algorithm','Error','N.features','Calc time','% of features']
df.head(1)
```

PROSTATE :

```python
df=prostate.sort_values(by=['Błąd','L.cech'])
df.columns=['Algorithm','Error','N.features','Calc time','% of features']
df.head(1)
```

#### 1.2 10 results with the smallest errors

```python
d=dlbcl.sort_values(by=['Błąd','L.cech']).head(10).reset_index(drop=True)
d.columns=['Algorithm','Error','N.features','Calc time','% of features']
p=prostate.sort_values(by=['Błąd','L.cech']).head(10).reset_index(drop=True)
p.columns=['Algorithm','Error','N.features','Calc time','% of features']
pd.concat([d,p],keys=['DLBCL','PROSTATE'],axis=1)
```

#### 1.3 Best results for each methods:

```python
d=dlbcl.sort_values(['Błąd','L.cech']).groupby('Algorytm').first().sort_values(['Błąd','L.cech']).reset_index()
d.columns=['Algorithm','Error','N.features','Calc time','% of features']
d.loc[d['Algorithm']=='Bez_selekcji','Algorithm']='No selection'

p=prostate.sort_values(['Błąd','L.cech']).groupby('Algorytm').first().sort_values(['Błąd','L.cech']).reset_index()
p.columns=['Algorithm','Error','N.features','Calc time','% of features']
p.loc[p['Algorithm']=='Bez_selekcji','Algorithm']='No selection'

pd.concat([d,p],keys=['DLBCL','PROSTATE'],axis=1)
```


Each algorithms for each datasets gave better results than classification with no feature selection (no selection). Results indicate a relationship between algorithm and dataset.

### 3. __ALGORITHMS__

#### 3.1 ANOVA

```python
d=dlbcl[dlbcl['Algorytm'].isin(['Anova','Bez_selekcji'])].sort_values(by=['Błąd','L.cech']).head(10).reset_index(drop=True)
d.columns=['Algorithm','Error','N.features','Calc time','% of features']
d.loc[d['Algorithm']=='Bez_selekcji','Algorithm']='No selection'

p=prostate[prostate['Algorytm'].isin(['Anova','Bez_selekcji'])].sort_values(by=['Błąd','L.cech']).head(10).reset_index(drop=True)
p.columns=['Algorithm','Error','N.features','Calc time','% of features']
p.loc[p['Algorithm']=='Bez_selekcji','Algorithm']='No selection'


pd.concat([d,p],keys=['DLBCL','PROSTATE'],axis=1)
```

Anova gives good or very good results (low Errors) no matter what datasets used for calculation.Drawback is relativly big dimension of selected features set.Algorithm is fast.  Best three results was generated with less than 15% of features.

#### 3.2 RFE

```python
d=dlbcl[dlbcl['Algorytm'].isin(['RFE','Bez_selekcji'])].sort_values(by=['Błąd','L.cech']).head(10).reset_index(drop=True)
d.columns=['Algorithm','Error','N.features','Calc time','% of features']
d.loc[d['Algorithm']=='Bez_selekcji','Algorithm']='No selection'

p=prostate[prostate['Algorytm'].isin(['RFE','Bez_selekcji'])].sort_values(by=['Błąd','L.cech']).head(10).reset_index(drop=True)
p.columns=['Algorithm','Error','N.features','Calc time','% of features']
p.loc[p['Algorithm']=='Bez_selekcji','Algorithm']='No selection'

pd.concat([d,p],keys=['DLBCL','PROSTATE'],axis=1)
```

RFE: results strongly depend on dataset. For DBCL set, except one case (50% od features selected), all results were worse than 'no selection' classification. For PROSTATE set all results were better then 'no selection' classification. Best RFE results for PROSTATE file beat other algorithms with low error and low number of selected features ('Błąd'=0,133, RFE : 2 features , ANOVA:214 features).The visible relationship between the number of features and the obtained result: inversely proportional in the case of a matched data set (PROSTATE, the deeper the algorithm goes, the better the result is) and almost directly proportional when the data set is not matched (DBCL: the deeper it goes the results is worse). Computationally expensive algorithm. Selection time two orders of magnitude greater than Anova and Decision Tree.

#### 3.3 DecisionTREE/RandomForest

```python
d=dlbcl[dlbcl['Algorytm'].isin(['Tree','Bez_selekcji'])].sort_values(by=['Błąd','L.cech']).head(10).reset_index(drop=True)
d.columns=['Algorithm','Error','N.features','Calc time','% of features']
d.loc[d['Algorithm']=='Bez_selekcji','Algorithm']='No selection'

p=prostate[prostate['Algorytm'].isin(['Tree','Bez_selekcji'])].sort_values(by=['Błąd','L.cech']).head(10).reset_index(drop=True)
p.columns=['Algorithm','Error','N.features','Calc time','% of features']
p.loc[p['Algorithm']=='Bez_selekcji','Algorithm']='No selection'

pd.concat([d,p],keys=['DLBCL','PROSTATE'],axis=1)
```

DecisionTree / RandomForest: As with RFE, the results obtained depend on the dataset. For DBCL: 8 of 10 best results are DecisionTree or RandomForest, for PROSTATE only 1 of 10 best results is  generated by DecisionTree. DecisionTree selects small feature subsets, usually producing the best results with larger values of % of features parameter, while RandomForest produces more numerous feature subsets.DT produces the highest error range of any algorithm (DBCL: 10.4% - 41.7%, PROSTATE: 15% -33 %) DT is a fast algorithm.

### __4.TIME__

#### 4.1 10 results with best computation time 

```python
d=dlbcl[~(dlbcl['Algorytm']=='Bez_selekcji')].sort_values(by=['Czas','Błąd']).head(10).reset_index(drop=True)
d.columns=['Algorithm','Error','N.features','Calc time','% of features']
p=prostate[~(prostate['Algorytm']=='Bez_selekcji')].sort_values(by=['Czas','Błąd']).head(10).reset_index(drop=True)
p.columns=['Algorithm','Error','N.features','Calc time','% of features']
pd.concat([d,p],keys=['DLBCL','PROSTATE'],axis=1)
```

#### 4.2 Selection mean time

```python
d=dlbcl.groupby(by='Algorytm').mean().Czas
d.index.name='Algorithm'
p=prostate.groupby(by='Algorytm').mean().Czas
p.index.name='Algorithm'
pd.concat([d,p],keys=['DLBCL','PROSTATE'],axis=1)

```

The selection time is closely related to the complexity of the algorithm and in principle does not depend on either the data set or the parameter % of features.
